### Define some helper functions

```python
import numpy as np
import tensorflow as tf
import random as python_random

def random_seed(seed = 123):
  
  # The below is necessary for starting Numpy generated random numbers
  # in a well-defined initial state.
  np.random.seed(seed)

  # The below is necessary for starting core Python generated random numbers
  # in a well-defined state.
  python_random.seed(seed)

  # The below set_seed() will make random number generation
  # in the TensorFlow backend have a well-defined initial state.
  # For further details, see:
  # https://www.tensorflow.org/api_docs/python/tf/random/set_seed
  tf.random.set_seed(seed)
```

Function below builds neural network
 - Caller should provide number of hidden layers, number of nodes per hidden layer, optimizer, batch size, number of epochs and multiplier for default learning rate (since default LR are different for different optimizers)
 - In the beginning of the proc we call random_seed() to reproduce the same results between runs
 - We use dropouts to help prevent overfitting. Dropout rate is also provided by the caller
 - We use training set to train NN and validation set for performance validation during training. Test set is used for evaluation of model performance when training is done

```python
def buildNeuralNetwork(n_layers, n_nodes, activation, loss_f, optimizer, 
                       lr_mult, batch, epochs, X, y, n_inputs, do_rate,
                       callbacks, verbose):
  random_seed()
  model = Sequential()
  #input layer
  model.add(ks.layers.InputLayer(input_shape = (n_inputs,)))
  #alternative input layer
  #model.add(Dense(n_nodes, input_shape=(n_inputs, ), kernel_initializer='he_normal', activation='relu'))
  #hidden layers  
  n_nodes_layer = n_nodes * 1.5
  for i in range(0, n_layers):
    n_nodes_layer /= 1.5    
    model.add(Dense(n_nodes_layer, activation=activation))
    if do_rate  > 0:
      model.add(Dropout(do_rate))
  #output layer
  model.add(Dense(1, activation='sigmoid'))

  sample_weight = np.ones(shape=(len(y),))
  sample_weight[y == 1] = 4

  val_sample_weight = np.ones(shape=(len(y_val_np),))
  val_sample_weight[y_val_np == 1] = 4

  opt = optimizer
  
  old_lr = optimizer.get_config()['learning_rate']
  new_lr = old_lr * lr_mult
  opt.learning_rate.assign(new_lr)
  print("Change LR from {} to {}".format(old_lr, opt.learning_rate))

  metrics = [
    ks.metrics.FalseNegatives(name="fn"),
    ks.metrics.FalsePositives(name="fp"),
    ks.metrics.TrueNegatives(name="tn"),
    ks.metrics.TruePositives(name="tp"),
    ks.metrics.Precision(name="precision"),
    ks.metrics.Recall(name="recall"),  
    ks.metrics.AUC(name='AUC'),
    ks.metrics.PrecisionAtRecall(0.7, name='PaR'),
 ]

  model.compile(optimizer=opt,
              loss=loss_f,
              metrics=metrics,
              )
  #history = model.fit(X, y, epochs=epochs, batch_size=batch, verbose=verbose,  
  #          sample_weight=sample_weight, validation_split=0.25, 
  #          callbacks = callbacks)
  history = model.fit(X, y, epochs=epochs, batch_size=batch, verbose=verbose, 
            sample_weight=sample_weight, 
            validation_data=(X_val_np, y_val_np, val_sample_weight),
            callbacks = callbacks)

  ##restore learning rate
  opt.learning_rate.assign(old_lr)
  return model, history
```

calc_custom_metric calculates custom metric for model performance estimation. It will be explained below

```python
recall_weight = 2
## Custom metric: see explanation below

def calc_custom_metric(y, y_pred):  
  tn, fp, fn, tp = skmetrics.confusion_matrix(y, y_pred).ravel()
  tpr = tp / (tp + fn)
  fpr = fp / (fp + tn)
  custom = (1 + recall_weight * tpr - fpr) / (recall_weight + 1) 
  return custom  

def get_model_metrics (model, x, y, threshold, out):
  y_pred = model.predict(x)
  y_pred = y_pred[:,0]
  step_f = lambda x : 0 if x < threshold else 1
  y_pred = np.array([step_f(xi) for xi in y_pred])


  recall, precision, f1, cm, custom = \
           skmetrics.recall_score(y, y_pred), \
           skmetrics.precision_score(y, y_pred),\
           skmetrics.f1_score(y, y_pred),\
           skmetrics.confusion_matrix(y, y_pred), \
           calc_custom_metric(y, y_pred)
  if (out):
    print ("Recall: {0:.2}, Precision: {1:.2}, f1:{2:.2}, custom metric:{3:.2}:".format(recall, precision, f1, custom))
  
  return recall, precision, f1, cm, custom
```

Functions in the below cell are used to calculate optimal threshold using custom metric

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx

def find_best_threshold(y, y_pred):
  fpr, tpr, thresholds = skmetrics.roc_curve(y, y_pred)

  optimal_idx = np.argmax((recall_weight * tpr - fpr))
  return thresholds[optimal_idx]
```

Some functions for data visualization 

```python
def make_confusion_matrix(cm,labels=[1, 0]):
    df_cm = pd.DataFrame(cm, index = [i for i in ["Actual - Loyal","Actual - Exited"]],
                  columns = [i for i in ['Predicted - Loyal','Predicted - Exited']])
    group_counts = ["{0:0.0f}".format(value) for value in
                cm.flatten()]
    group_percentages = ["{0:.2%}".format(value) for value in
                         cm.flatten()/np.sum(cm)]
    labels = [f"{v1}\n{v2}" for v1, v2 in
              zip(group_counts,group_percentages)]
    labels = np.asarray(labels).reshape(2,2)
    plt.figure(figsize = (10,7))
    sns.heatmap(df_cm, annot=labels,fmt='')
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
def plot_history(history,  keys):
  for key in keys:
    plt.plot(history.history[key])
    plt.plot(history.history['val_'+key])
    plt.ylabel(key)
    plt.xlabel('epoch')
    plt.legend(['train','test'], loc='upper left')
    plt.show()
```

```python
def plot_roc_auc(model, X, y, best_th=-1):
  y_pred=model.predict(X)[:,0]
  fpr_keras, tpr_keras, thresholds_keras = skmetrics.roc_curve(y, y_pred)
  auc_keras = skmetrics.auc(fpr_keras, tpr_keras)
  plt.figure(1)
  plt.plot([0, 1], [0, 1], 'k--')
  plt.plot(fpr_keras, tpr_keras, label='Keras (area = {:.3f})'.format(auc_keras))
  plt.xlabel('False positive rate')
  plt.ylabel('True positive rate')
  plt.title('ROC curve')
  if (best_th >= 0) :
    best_idx = find_nearest(thresholds_keras, best_th)
    best_x = fpr_keras[best_idx]
    best_y = tpr_keras[best_idx]
    plt.plot(best_x, best_y, 'ro', label='Best score')
  plt.legend(loc='best')
  plt.show()
  plt.show()

  
```

```python
def plot_precision_recall(model, X, y, best_th):
  y_pred=model.predict(X)[:,0]
  pr_keras, rc_keras, thresholds_keras = skmetrics.precision_recall_curve(y, y_pred)
  plt.figure(1)
  plt.plot(pr_keras, rc_keras, label='Precision vs Recall')
  plt.xlabel('Precision')
  plt.ylabel('Recall')
  plt.title('Precision-Recall curve')
  if (best_th >= 0) :
    best_idx = find_nearest(thresholds_keras, best_th)
    best_x = pr_keras[best_idx]
    best_y = rc_keras[best_idx]
    plt.plot(best_x, best_y, 'ro', label='Best score')
  plt.legend(loc='best')
  plt.show()
  plt.show()
```

Threshold can be changed to  make model performance better. Before finding the optimal threshold we should decide what metric we are going to use

Loosing a customer has a big impact on bank and we should identify as many of exiting customers as possible, which means True Positive Rate a.k.a. Recall should be maximized. But NN can maximize recall by simply identify all the customers as Exiting. To prevent that we should penalize our metric for every loyal customer identified as exiting. 
 \begin{equation}
metric= w_{recall} * TPR - FPR
\end{equation}
$w_{recall}$ is a weight of an Exiting customer compared to the weight of a loyal customer. For example, if identifying of exiting customer is twice as important as identifying a loyal customer, then $w_{recall} = 2$.
This metric is in range of [-1, $w_{recall} - 1$]. In order to place it in [0,1] range we should modify metric as follows:
 \begin{equation}
metric= \frac{(1 + w_{recall} * TPR - FPR)}{1+w_{recall}}
\end{equation}
This metric is calculated by finction *calc_custom_metric*.
Custom Test data score for the original NN is 0.74

```python
y_pred=model.predict(X_val_np)[:,0]
pr, rc, thresholds_keras = skmetrics.precision_recall_curve(y_val_np, y_pred)

#f1 = 2 * (pr * rc) / (pr + rc)
#idx = find_nearest(f1, f1.max())

new_th = find_best_threshold(y_val_np, y_pred)
idx = find_nearest(thresholds_keras, new_th)

print ("{} Precision={}, Recall={}, Threshold={}".format(idx, pr[idx], \
                  rc[idx], thresholds_keras[idx]))
```

## Early stopping

As mentioned earlier, loss function for validation set stops decreasing and start increasing after certain epoch. We can try to catch this moment and stop training NN to avoid overfitting

```python
early_stop_cb = ks.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0,
    patience=20,
    verbose=0,
    mode="min",
    baseline=None,
    restore_best_weights=True)

callbacks = [early_stop_cb,]
```

```python
model,history = buildNeuralNetwork(n_layers, n_nodes, activation, loss_f, optimizer, mult, batch, 
                   epochs, X_train_np, y_train_np, 
                   n_inputs, do_rate, callbacks, verbose)
```

## Learning rate adjustment

As mentioned earlier, loss function demonstrates oscilations which is a sign of leaning rate being too high. We can try to start reducing learning rate after certain epoch to supress oscilations

```python
from tensorflow.math import  exp

def lr_scheduler(epoch, lr):
  if epoch < 10:
    return lr
  elif epoch == 10:
    return lr * 0.5
  else:
    return lr * exp(-0.05)
```

# Appendix: Using simulated annealing technique for NN training

As mentioned earlier, some kind of golbal optimization can be tried for NN training. Here is my attempt to implement global optimization algorithm called Simulated Annealing. There are many resources on the Web explaining Simulated Annealing algorithm. For example, this one:
https://www.theprojectspot.com/tutorial-post/simulated-annealing-algorithm-for-beginners/6

Gradient descent is often explained as a behavior of a solid ball going down from the hill to the bottom of the valley. To explany a concept of global optimiztion we can use a model of elastic ball that can also go downhill, but it can also bounce. If there are multiple hills and valleys in the multidimentional space, the elastic ball will be able to bounce out of the shallow valley and find a deeper, more optimal, one. 

In the beginning the bouncing energy of the ball is high and it's mostly bouncing. Then it combines bouncing and gradient descent behaviour. At the end of the optimization it only descends.



```python
indent = 0
mutation_ratio = 0.1
mutator = (lambda x: x if np.random.rand() > mutation_ratio else 0.0)

def mutate(mutation_ratio, weights):
  global indent 

  result = []

  new_weights = weights
  indent += 1
  if type(weights) in [list]:
    #print('.'* indent, type(weights), len(weights))
    for w in weights:
      new_w = mutate(mutation_ratio, w)
      result.append(new_w)
  elif type(weights) in [np.ndarray]:
    #print('.'* indent, type(weights), len(weights), weights.shape)
    if len(weights.shape) == 1:
      #print('---',weights)
      m = map(mutator, weights)
      result=np.array(list(m))
      #print('.'* indent, 'result 1:',  type(result), len(result), result.shape)
      #print ('+++',result)
    else:
      result = []
      for w in weights:
        new_w = mutate(mutation_ratio, w)
        result.append(new_w)
      result=np.array(result)
      #print('.'* indent, 'result 2:',  type(result), len(result), result.shape)
  else:
    1
  indent -= 1
  return result
```

The code above implements a function mutate() which is a simulation of a single bounce. I changes randomly zeros 10% of model weights, which leads (in most cases) to the increas of model loss function 

The code below implements a custom callback class with simulated annealing functionality. Parameter T is a "Temperature" parameter; it can also be explained as ball's bouncing energy. The higher T, the higher is a probability to bounce. 

In the beginning of every epoch the ball makes an attempt to bounce, i.e. we zero 10% of the model weights. If new loss function is lower than loss function of previous epoch, we accept this bounce. If new loss function is higher, we calculate a probabity of that bounce and compare with a rundom number. If probabity is higher than a random number, we accept the bounce even if it leads to the increase of loss function. Othervise, we decline bounce and perform gradient step.

Parameter T decrease over time to model the reduction of bouncing energy. As T decreses, the probability of a successfull bounce decreases as well. T decreases by 2% after each successfull bounce or every 10th epoch whichever happens first.

During the training we save weights for the best val_loss and restore them at the end of training

```python
class SaCallback(ks.callbacks.Callback):
    def on_train_begin(self, logs=None):
        self.last_loss = 0.0
        self.T=50.0        
        self.epoch_per_temp = 10
        self.temp_change_rate = 0.025
        self.sample_weight = np.ones(shape=(len(y_train_np),))
        self.sample_weight[y_train_np == 1] = 4

        self.val_sample_weight = np.ones(shape=(len(y_val_np),))
        self.val_sample_weight[y_val_np == 1] = 4
        
        self.min_delta = 1e-3

        self.prev_bounce_epoch = 0
        self.prev_improvement_epoch = 0
        self.patience = 15
        self.best_loss = 1e300
        self.best_weights = None

        self.descend_after_bounce = 1

        return

    def on_train_end(self, logs=None):
        print ("T at the end of training is {0:.2}".format(self.T))
        self.model.set_weights(self.best_weights)
        return

    def on_epoch_begin(self, epoch, logs=None):
      if epoch % 100 == 0:
        print ("Epoch # {}".format(epoch))
      if(self.can_mutate(epoch)):
        self.change_model()
        action = 'bounce'
        change, change_loss = self.accept_chage()
        if change == 0:
          action = 'descend'
          self.model.set_weights(self.save_weights)
          
        #print("Begin {0} val_loss = {1:.4}: {2} T:{3:.2} changed loss: {4:.4}".format(epoch, self.last_loss, action, self.T, change_loss))
        if (action == 'bounce'):
          self.prev_bounce_epoch = epoch
          self.T = self.T * (1 - self.temp_change_rate)

      return

    def on_epoch_end(self, epoch, logs=None):
      if logs['val_loss'] < self.last_loss:
        self.prev_improvement_epoch = epoch

      self.last_loss = logs['val_loss']
      if (epoch % self.epoch_per_temp == self.epoch_per_temp - 1):      
        self.T = self.T * (1 - self.temp_change_rate)
      
      if (epoch > self.patience and (epoch - self.prev_bounce_epoch) > self.patience):
        lr = (K.eval(self.model.optimizer.lr))
        lr *= 0.999
        if lr < 1e-5:
          lr = 1e-5
        #print ("Change LR to {}".format(lr))
        self.model.optimizer.learning_rate.assign(lr)

      if self.last_loss < self.best_loss:
        self.best_loss = self.last_loss
        self.best_weights = self.model.get_weights()
        
      #if (epoch - self.prev_improvement_epoch > self.patience):
      #  self.model.stop_training = True
      return

    def on_test_begin(self, logs=None):
        return

    def on_test_end(self, logs=None):
        return

    def on_predict_begin(self, logs
                         =None):
        return

    def on_predict_end(self, logs=None):
        return

    def on_train_batch_begin(self, batch, logs=None):
        return

    def on_train_batch_end(self, batch, logs=None):
        return

    def on_test_batch_begin(self, batch, logs=None):
        return

    def on_test_batch_end(self, batch, logs=None):
        return

    def on_predict_batch_begin(self, batch, logs=None):
        return

    def on_predict_batch_end(self, batch, logs=None):
        return

    def can_mutate(self, epoch):
      if (epoch - self.prev_bounce_epoch < self.descend_after_bounce):
        return 0
      return 1

    def accept_chage(self):
      y_pred = self.model.predict(X_val_np)
      loss = self.model.loss(y_val_np, y_pred, self.val_sample_weight)
      dL = loss - self.last_loss

      #not interested in very low bounce
      if (abs(dL) < self.min_delta):
        return 0, loss

      probability = exp(-dL/self.T)
      if (probability >= np.random.rand()):
        return 1, loss

      return 0, loss

    def change_model(self):
      self.save_weights = self.model.get_weights()
      new_weights = mutate(mutation_ratio, self.save_weights)
      self.model.set_weights(new_weights)
      return
```
